```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import joblib

print("Libraries imported successfully!")
```

```python
import pandas as pd
import numpy as np

# Número de estudiantes
n_students = 1000

# Establecer una semilla para reproducibilidad
np.random.seed(42)

# Generar variables independientes (todas numéricas: 1 para sí, 0 para no en binarias)

edad = np.random.randint(18, 30, n_students) # Edad entre 18 y 29
promedio = np.random.normal(3.0, 0.7, n_students) # Promedio alrededor de 3.0
promedio = np.clip(promedio, 0.0, 4.0) # Asegurar que esté entre 0.0 y 4.0

asistencia = np.random.beta(a=10, b=2, size=n_students) # Valores más altos indican mejor asistencia (cerca de 1)
horas_estudio = np.random.randint(1, 40, n_students) # Horas de estudio por semana
uso_plataforma = np.random.beta(a=5, b=5, size=n_students) # Valores más altos indican más uso de la plataforma
materias_perdidas = np.random.poisson(lam=0.8, size=n_students) # Mayoría 0 o 1, algunos hasta 5
materias_perdidas = np.clip(materias_perdidas, 0, 5)

nivel_socioeconomico = np.random.beta(a=2, b=2, size=n_students) # 0 (bajo) a 1 (alto)
trabaja = np.random.choice([0, 1], size=n_students, p=[0.6, 0.4]) # 60% no trabaja, 40% sí
acceso_internet = np.random.choice([0, 1], size=n_students, p=[0.1, 0.9]) # 10% sin internet, 90% con internet

# Crear un DataFrame para facilitar la manipulación
data = pd.DataFrame({
    'edad': edad,
    'promedio': promedio,
    'asistencia': asistencia,
    'horas_estudio': horas_estudio,
    'uso_plataforma': uso_plataforma,
    'materias_perdidas': materias_perdidas,
    'nivel_socioeconomico': nivel_socioeconomico,
    'trabaja': trabaja,
    'acceso_internet': acceso_internet
})

# Generar la variable 'deserta' (dropout) basada en patrones realistas
# Se asignan pesos a las variables que influyen en la probabilidad de deserción
# Un 'coeficiente' positivo aumenta la probabilidad de deserción, negativo la disminuye

# Inicializar una probabilidad base de deserción
prob_desercion = np.full(n_students, 0.10) # 10% de probabilidad base

# Ajustar la probabilidad de deserción según las variables
prob_desercion += (data['edad'] - data['edad'].mean()) / data['edad'].std() * 0.02 # Edad más alta ligeramente aumenta
prob_desercion -= (data['promedio'] - data['promedio'].mean()) / data['promedio'].std() * 0.08 # Promedio bajo aumenta
prob_desercion -= (data['asistencia'] - data['asistencia'].mean()) / data['asistencia'].std() * 0.05 # Baja asistencia aumenta
prob_desercion -= (data['horas_estudio'] - data['horas_estudio'].mean()) / data['horas_estudio'].std() * 0.03 # Bajas horas de estudio aumentan
prob_desercion -= (data['uso_plataforma'] - data['uso_plataforma'].mean()) / data['uso_plataforma'].std() * 0.04 # Bajo uso de plataforma aumenta
prob_desercion += (data['materias_perdidas'] - data['materias_perdidas'].mean()) / data['materias_perdidas'].std() * 0.10 # Más materias perdidas fuertemente aumenta
prob_desercion -= (data['nivel_socioeconomico'] - data['nivel_socioeconomico'].mean()) / data['nivel_socioeconomico'].std() * 0.07 # Nivel socioeconómico bajo aumenta
prob_desercion += data['trabaja'] * 0.10 # Trabajar aumenta la probabilidad
prob_desercion += (1 - data['acceso_internet']) * 0.12 # No tener acceso a internet fuertemente aumenta

# Asegurar que la probabilidad esté en el rango [0, 1]
prob_desercion = np.clip(prob_desercion, 0.01, 0.90) # Mínimo 1%, Máximo 90% de chance de deserción

# Generar la variable 'deserta' (1 si deserta, 0 si no) basada en la probabilidad
data['deserta'] = (np.random.rand(n_students) < prob_desercion).astype(int)

# Mostrar las primeras 5 filas del dataset y la distribución de la variable 'deserta'
display(data.head())
print("\nDistribución de la variable 'deserta':")
display(data['deserta'].value_counts(normalize=True))
```

### 6. Data Splitting: Features, Target, Training, and Test Sets

```python
# 1. Separar variables predictoras (X) y variable objetivo (y)
X = data.drop('deserta', axis=1)
y = data['deserta']

print(f"Shape of features (X): {X.shape}")
print(f"Shape of target (y): {y.shape}")
```

```python
# 2. y 3. Dividir datos en entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")

print("\nDistribution of 'deserta' in y_train:")
display(y_train.value_counts(normalize=True))
print("\nDistribution of 'deserta' in y_test:")
display(y_test.value_counts(normalize=True))
```

### 7. Model Training: RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

# Inicializar el modelo RandomForestClassifier
model = RandomForestClassifier(random_state=42)

# Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

print("Modelo RandomForestClassifier entrenado exitosamente!")
```

### 8. Making Predictions

```python
# Realizar predicciones sobre los datos de prueba
y_pred = model.predict(X_test)

print("Predicciones realizadas en X_test.")
# Opcional: mostrar las primeras 5 predicciones
display(y_pred[:5])
```

### 9. Model Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Mostrar el reporte de clasificación
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Mostrar la matriz de confusión
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
display(pd.DataFrame(cm, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1']))

plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

### 10. Feature Importance

```python
importances = model.feature_importances_
feature_names = X_train.columns

forest_importances = pd.Series(importances, index=feature_names)

fig, ax = plt.subplots(figsize=(10, 6))
forest_importances.sort_values(ascending=False).plot.bar(ax=ax)
ax.set_title("Importancia de las Características del Modelo")
ax.set_ylabel("Disminución media de la impureza")
plt.tight_layout()
plt.show()

display(forest_importances.sort_values(ascending=False))
```

## 12. Streamlit Application for Dropout Prediction

Now, let's create a simple web application using Streamlit to interact with our trained model. This app will allow users to input student data and get a real-time prediction on the likelihood of dropout.

First, we'll write the Streamlit application code to a Python file (`app.py`), and then we'll run it using `pyngrok` to expose it via a public URL.

```python
import streamlit as st
import joblib
import numpy as np
import pandas as pd

# Guardar el contenido del archivo app.py
# Este código se ejecutará cuando Streamlit levante el archivo.
app_py_content = """
import streamlit as st
import joblib
import numpy as np
import pandas as pd

# --- 1. Cargar el Modelo Entrenado ---
# Asegúrate de que 'modelo_desercion.pkl' esté en el mismo directorio
# o proporciona la ruta completa.
try:
    model = joblib.load('modelo_desercion.pkl')
    st.success("Modelo 'modelo_desercion.pkl' cargado exitosamente!")
except FileNotFoundError:
    st.error("Error: El archivo 'modelo_desercion.pkl' no fue encontrado. Asegúrate de que esté en la ruta correcta.")
    st.stop() # Detener la ejecución si el modelo no se carga

# --- 2. Título y Descripción de la Aplicación ---
st.set_page_config(page_title="Predicción de Deserción Estudiantil")
st.title("🎓 Sistema de Predicción de Deserción Estudiantil")
st.markdown("Utiliza los siguientes parámetros para predecir si un estudiante tiene riesgo de desertar.")

# --- 3. Sidebar para Inputs (o Layout de Columnas) ---
st.sidebar.header("Parámetros del Estudiante")

# Inputs para cada característica del modelo
edad = st.sidebar.slider("Edad", 18, 30, 22)
promedio = st.sidebar.slider("Promedio Académico (0.0 - 4.0)", 0.0, 4.0, 2.5, 0.1)
asistencia = st.sidebar.slider("Asistencia (0.0 - 1.0, 1.0 es alta)", 0.0, 1.0, 0.85, 0.05)
horas_estudio = st.sidebar.slider("Horas de Estudio por Semana", 1, 40, 15)
uso_plataforma = st.sidebar.slider("Uso de Plataforma Educativa (0.0 - 1.0, 1.0 es alto)", 0.0, 1.0, 0.7, 0.05)
materias_perdidas = st.sidebar.slider("Número de Materias Perdidas", 0, 5, 0)
nivel_socioeconomico = st.sidebar.slider("Nivel Socioeconómico (0.0 - 1.0, 1.0 es alto)", 0.0, 1.0, 0.5, 0.05)
trabaja = st.sidebar.selectbox("¿El estudiante trabaja?", [0, 1], format_func=lambda x: "Sí" if x == 1 else "No")
acceso_internet = st.sidebar.selectbox("¿Tiene acceso a Internet?", [0, 1], format_func=lambda x: "Sí" if x == 1 else "No")

# --- 4. Preparar Datos para Predicción ---
input_data = pd.DataFrame([{
    'edad': edad,
    'promedio': promedio,
    'asistencia': asistencia,
    'horas_estudio': horas_estudio,
    'uso_plataforma': uso_plataforma,
    'materias_perdidas': materias_perdidas,
    'nivel_socioeconomico': nivel_socioeconomico,
    'trabaja': trabaja,
    'acceso_internet': acceso_internet
}])

# --- 5. Realizar Predicción ---
st.subheader("Resultado de la Predicción")
if st.button("Predecir Deserción"):
    try:
        prediction = model.predict(input_data)[0]
        prediction_proba = model.predict_proba(input_data)[:, 1][0] # Probabilidad de deserción (clase 1)

        if prediction == 1:
            st.error(f"🚨 **¡ALERTA! Alto riesgo de deserción.** (Probabilidad: {prediction_proba:.2%})")
            st.markdown("Se recomienda intervención y apoyo inmediato. Contactar con el estudiante y ofrecer recursos como asesoramiento académico, apoyo financiero o tutorías.")
        else:
            st.success(f"✅ **Bajo riesgo de deserción.** (Probabilidad: {prediction_proba:.2%})")
            st.markdown("El estudiante parece estar en buen camino. Continuar monitoreo regular.")

    except Exception as e:
        st.error(f"Ocurrió un error durante la predicción: {e}")

# --- 6. Información Adicional (Opcional) ---
st.markdown("---")
st.info("Esta herramienta es solo para fines predictivos y debe usarse como complemento a la evaluación profesional.")


"""

with open('app.py', 'w') as f:
    f.write(app_py_content)

print("Archivo 'app.py' creado exitosamente!")
```
